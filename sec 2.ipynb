```python
import networkx as nx
```

```python
 g= nx.Graph()
```

```python
g.add_node('a')
```

```python
nodes = ['a','b','c','d']
```

```python
g.add_nodes_from(nodes)
```

```python
g.add_edge('a','b')
```

```python
edges=[('a','b'),('b','c'),('a','c'),('c','d')]
```

```python
g.add_edges_from(edges)
```

```python
nx.draw(g,with_labels=True,node_color='green',node_size=1600)
```

```python
D = nx.DiGraph()
D.add_edges_from([(1,2),(2,3),(1,3),(3,4)])
nx.draw(D,with_labels=True,node_size=1600,node_color="brown")
```

```python
d = nx.Graph()
d.add_edges_from([(1,2),(1,3),(2,3),(4,5),(4,3),(5,3)])
nx.draw(d,with_labels=True,node_color="brown",node_size=1600)
```
